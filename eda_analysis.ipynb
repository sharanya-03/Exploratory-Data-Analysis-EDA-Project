##Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option('display.max_columns', None)

# Plot settings
plt.rcParams['figure.figsize'] = (8,5)

print("Libraries imported successfully.")
```

```python
## Loading the Dataset

In this step, we load the dataset into a Pandas DataFrame and inspect the first few records.
```

```python
df = pd.read_csv("Superstore.csv")

df.head()
```

## Dataset Overview

Understanding the size, structure, and data types of the dataset.

```python
print("Shape of Dataset:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nDataset Information:")
df.info()
```

## Statistical Summary

Generate descriptive statistics to understand the distribution of numerical variables.

```python
df.describe()
```

## Missing Values Analysis

Check whether the dataset contains missing values.

```python
df.isnull().sum()
```

## Data Cleaning

Handling missing values and removing duplicate records.

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Fill missing numerical values with mean
df.fillna(df.mean(numeric_only=True), inplace=True)

print("Data cleaning completed.")
```

## Distribution Analysis

Visualizing the distribution of numerical variables.

```python
plt.figure(figsize=(8,5))

df['Age'].hist(bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

## Outlier Detection

Box plots help identify potential outliers in the dataset.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Salary'])

plt.title("Salary Distribution")

plt.show()
```

## Categorical Variable Analysis

Analyze the frequency distribution of categorical variables.

```python
df['Department'].value_counts().plot(kind='bar')

plt.title("Department Distribution")
plt.xlabel("Department")
plt.ylabel("Count")

plt.show()
```

## Correlation Analysis

Understanding relationships among numerical variables using a correlation matrix.

```python
plt.figure(figsize=(10,6))

correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

## Relationship Between Variables

Scatter plots help visualize relationships between two numerical variables.

```python
sns.scatterplot(
    x='Experience',
    y='Salary',
    data=df
)

plt.title("Experience vs Salary")

plt.show()
```

## Key Insights

### Insight 1
Most observations are concentrated within a specific range, indicating a normal distribution pattern.

### Insight 2
Several outliers were identified in the salary variable.

### Insight 3
A positive relationship exists between experience and salary.

### Insight 4
Certain variables exhibit strong correlations, suggesting they may influence each other.
